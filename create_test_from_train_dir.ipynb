```
import os
import csv
```

```
PERCENTAGE_TEST_DATA = 33
CURRENT_DIR = '/content/gdrive/My Drive/CITS Color/test'
TRAIN_DIR = CURRENT_DIR
TEST_DIR = CURRENT_DIR+'/validation'
CSV_FILE = "/CITS_Color_test.csv"
speciesList = []
speciesValue = {}
speciesData={}
preparingTestData = {}
fixTestData = {}
newArrayTemp = []
```

```
def newArray(p, fill):
    newArrayTemp = []
    for i in range(0, p):
        if (i == fill):
            newArrayTemp.append(1)
        else:
            newArrayTemp.append(0)
    return newArrayTemp

def distinctList(theList):
    return list(set(theList))

def label_img(img_label):
    species_label=speciesData[img_label]
    if species_label in speciesValue.keys():
        return speciesValue[species_label]
    else:
        return("Unknown Species")


f = open(CURRENT_DIR+CSV_FILE)
csv_f = csv.reader(f)
for index, row in enumerate(csv_f):
  if index != 0:
      speciesList.append(row[2])
      speciesData[row[3]] = row[2]


speciesList = distinctList(speciesList)


for index1, species in enumerate(speciesList):
    f_test = open(CURRENT_DIR + CSV_FILE)
    csv_test = csv.reader(f_test)
    counter = 0
    for index2, row in enumerate(csv_test):
        if index2 != 0:
            if row[2] == species:
                counter = counter +1
    print (counter*PERCENTAGE_TEST_DATA/100),"-",species
    testdata = {}
    testdata['counter'] = counter
    testdata['test_data'] = counter*PERCENTAGE_TEST_DATA/100
    preparingTestData[species]=testdata


for index, species in enumerate(speciesList):
    f_test = open(CURRENT_DIR + CSV_FILE)
    csv_test = csv.reader(f_test)
    counter = 0
    tempNamaFileTestData = []
    for nameFile in speciesData:
        nameSpecies = speciesData[nameFile]
        if nameSpecies == species:
            counter = counter + 1
            if counter <= preparingTestData[species]['test_data']:
                print counter," - ",species, " - ", preparingTestData[species]['test_data'], " - ", nameFile
                tempNamaFileTestData.append(nameFile)
    fixTestData[species]=tempNamaFileTestData
```

```
print fixTestData

for species in fixTestData:
    for fileNameTest in fixTestData[species]:
        if not (os.path.exists(TEST_DIR+'/'+fileNameTest.split("/")[0])):
            os.makedirs(TEST_DIR+'/'+fileNameTest.split("/")[0])
        print "move "+TRAIN_DIR+'/'+fileNameTest+ " to "+TEST_DIR+'/'+fileNameTest
        os.rename(TRAIN_DIR+'/'+fileNameTest, TEST_DIR+'/'+fileNameTest)


print "Sukses membagi training data ke testing data"
print preparingTestData
```
