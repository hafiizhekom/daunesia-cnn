```
import os
import sys
from os import walk
import csv

```

```
folder = '/content/gdrive/My Drive/CITS Color/test'
namaCsv = '/content/gdrive/My Drive/CITS Color/CITS_Color_test.csv'

if namaCsv[-4:] != ".csv":
    namaCsv = namaCsv+".csv"

theData =  os.listdir(folder)
theData.sort()
number = 1
with open(namaCsv, 'wb') as myfile:
    wr = csv.writer(myfile)
    wr.writerow(["nomor", "nama_file", "spesies", "path"])
    for folderData in theData:
        folderDataComplete = str(folder)+ "/" + str(folderData)
        for (dirpath, dirnames, filenames) in walk(folderDataComplete):
            for x in filenames:
                print number, " - ", x, " - ", folderData, " - ", folderData+"/"+x
                wr.writerow([number, x, folderData, folderData+"/"+x])
                number=number+1

```
